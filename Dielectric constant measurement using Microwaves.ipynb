```python
def tanxx(t,x):
    t = t*10**(-3)
    x = x * 10**(-2)
    num = betago*(t+x)
    den = betago*t
    
    return np.tan(num)/den
```

```python
def f(x,rhs):
    return np.tan(x)/x - rhs

def bisection(f,rhs,a,b,N):
    
    if f(a,rhs)*f(b,rhs) >= 0:
        print("Bisection method fails.")
        return None
    a_n = a
    b_n = b
    for n in range(1,N+1):
        m_n = (a_n + b_n)/2
        f_m_n = f(m_n,rhs)
        
        if f(a_n,rhs)*f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n,rhs)*f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            print("Found exact solution.")
            return m_n
        else:
            print("Bisection method fails.")
            return None
    return (a_n + b_n)/2
```

```python
plt.figure(figsize=(7,5))
x = np.linspace(0.1,1.4,100)
y =np.tan(x)/x
z = np.ones(100)*tanxx(t1,xshift1)

sol1 = bisection(f,tanxx(t1,xshift1),1.1,1.4,50)
plt.plot(x,y)
plt.plot(x,z,label="y = %.2f"%(tanxx(t1,xshift1)))
plt.axvline(sol1,ls='--',c='k',label="sol = {}".format(sol1))
#plt.grid()
plt.legend(prop={'size': 14})
plt.xlabel("X",fontsize=12)
plt.ylabel("tanX/X",fontsize=12)
plt.savefig("tanxxt1.pdf")
```
